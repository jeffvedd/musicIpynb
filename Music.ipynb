````python
# ETAPA 4 — Localizar músicas do dataset
import os, glob, shutil

NOME_DATASET = 'meus-audios'  # ← edite com o nome do seu dataset Kaggle

AUDIO_DIR  = '/kaggle/working/audios'
OUTPUT_DIR = '/kaggle/working/musicgen_lora'
os.makedirs(AUDIO_DIR,  exist_ok=True)
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Busca arquivos de áudio
extensoes = ['*.mp3', '*.wav', '*.flac', '*.ogg']
audios = []
for ext in extensoes:
    audios += glob.glob(f'/kaggle/input/{NOME_DATASET}/**/{ext}', recursive=True)
    audios += glob.glob(f'/kaggle/input/{NOME_DATASET}/{ext}')
audios = sorted(set(audios))

# Copia para pasta de trabalho
for a in audios:
    shutil.copy2(a, AUDIO_DIR)

arquivos = sorted(os.listdir(AUDIO_DIR))
print(f'🎵 {len(arquivos)} arquivo(s) de áudio encontrado(s):')
for i, a in enumerate(arquivos):
    tam = os.path.getsize(f'{AUDIO_DIR}/{a}') / 1e6
    print(f'  [{i}] {a}  ({tam:.1f} MB)')

if not arquivos:
    print('⚠️  Nenhum áudio encontrado! Verifique NOME_DATASET e se o dataset foi adicionado.')
    print('\nDatasets disponíveis:')
    for d in os.listdir('/kaggle/input'):
        print(f'  /kaggle/input/{d}/')
````

````python
# ETAPA 5 — Escrever prompts para cada música
# Um prompt por arquivo, mesma ordem da lista acima
PROMPTS = [
    'upbeat electronic dance music with heavy bass and synth leads',  # áudio 0
    'calm acoustic guitar fingerpicking in the style of bossa nova',  # áudio 1
    'epic orchestral music with strings and brass, cinematic feel',   # áudio 2
    # adicione mais conforme seus arquivos...
]

if len(PROMPTS) != len(arquivos):
    print(f'⚠️  {len(arquivos)} áudio(s) mas {len(PROMPTS)} prompt(s) — ajuste a lista!')
else:
    print('✅ Mapeamento áudio → prompt:')
    for a, p in zip(arquivos, PROMPTS):
        print(f'  🎵 {a}')
        print(f'     → {p}')
````

````python
# ETAPA 6 — Carregar modelo MusicGen e configurar LoRA
import torch
from transformers import AutoProcessor, MusicgenForConditionalGeneration
from peft import LoraConfig, get_peft_model, TaskType

MODEL_ID   = 'facebook/musicgen-small'  # small=300M roda na T4
LORA_RANK  = 8
DEVICE     = 'cuda' if torch.cuda.is_available() else 'cpu'

print('Carregando modelo MusicGen...')
processor = AutoProcessor.from_pretrained(MODEL_ID)
model = MusicgenForConditionalGeneration.from_pretrained(
    MODEL_ID,
    torch_dtype=torch.float16
)

# Configura LoRA no decoder do MusicGen
lora_config = LoraConfig(
    r=LORA_RANK,
    lora_alpha=LORA_RANK * 2,
    target_modules=['q_proj', 'k_proj', 'v_proj', 'out_proj'],
    lora_dropout=0.05,
    bias='none',
)

model.decoder = get_peft_model(model.decoder, lora_config)
model.decoder.print_trainable_parameters()
model = model.to(DEVICE)

print(f'\n✅ Modelo carregado em {DEVICE}!')
````

````python
# ETAPA 7 — Preparar dataset de áudio
import torch
import librosa
import numpy as np
from torch.utils.data import Dataset, DataLoader

SAMPLE_RATE = 32000  # MusicGen usa 32kHz
DURACAO_SEG = 10     # segundos por amostra

class MusicDataset(Dataset):
    def __init__(self, audio_dir, prompts, processor, sr=32000, duracao=10):
        self.samples = []
        arquivos = sorted([f for f in os.listdir(audio_dir) if not f.startswith('.')])
        for arquivo, prompt in zip(arquivos, prompts):
            caminho = os.path.join(audio_dir, arquivo)
            try:
                audio, _ = librosa.load(caminho, sr=sr, mono=True, duration=duracao)
                if len(audio) < sr * 2:  # mínimo 2 segundos
                    print(f'  ⚠️  {arquivo} muito curto, pulando')
                    continue
                # Padding para duração fixa
                target_len = sr * duracao
                if len(audio) < target_len:
                    audio = np.pad(audio, (0, target_len - len(audio)))
                else:
                    audio = audio[:target_len]
                self.samples.append({'audio': audio, 'prompt': prompt, 'nome': arquivo})
                print(f'  ✅ {arquivo} carregado')
            except Exception as e:
                print(f'  ❌ Erro em {arquivo}: {e}')
        self.processor = processor
        self.sr = sr

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        s = self.samples[idx]
        inputs = self.processor(
            text=[s['prompt']],
            audio=s['audio'],
            sampling_rate=self.sr,
            return_tensors='pt',
            padding=True
        )
        return inputs

print('Carregando dataset de áudio...')
dataset = MusicDataset(AUDIO_DIR, PROMPTS, processor, SAMPLE_RATE, DURACAO_SEG)
print(f'\n✅ Dataset pronto com {len(dataset)} amostra(s)!')
````

````python
# ETAPA 8 — Fine-Tuning LoRA
import torch
from torch.optim import AdamW
from torch.optim.lr_scheduler import CosineAnnealingLR

EPOCHS     = 5      # épocas de treino
LR         = 5e-5   # learning rate
GRAD_ACCUM = 4      # acumular gradiente (simula batch maior)

optimizer  = AdamW(model.parameters(), lr=LR, weight_decay=0.01)
scheduler  = CosineAnnealingLR(optimizer, T_max=EPOCHS)

model.train()
torch.cuda.empty_cache()

print(f'▶️  Iniciando treino — {EPOCHS} épocas, {len(dataset)} amostra(s)\n')

for epoch in range(EPOCHS):
    total_loss = 0
    optimizer.zero_grad()

    for step, batch in enumerate(dataset):
        try:
            input_ids      = batch['input_ids'].squeeze(0).to(DEVICE)
            attention_mask = batch['attention_mask'].squeeze(0).to(DEVICE)

            with torch.autocast('cuda', dtype=torch.float16):
                outputs = model(
                    input_ids=input_ids,
                    attention_mask=attention_mask,
                    labels=input_ids
                )
                loss = outputs.loss / GRAD_ACCUM

            loss.backward()
            total_loss += loss.item() * GRAD_ACCUM

            if (step + 1) % GRAD_ACCUM == 0 or step == len(dataset) - 1:
                torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
                optimizer.step()
                optimizer.zero_grad()

        except RuntimeError as e:
            if 'out of memory' in str(e):
                print(f'  ⚠️  OOM no step {step}, pulando...')
                torch.cuda.empty_cache()
                optimizer.zero_grad()
            else:
                raise e

    scheduler.step()
    media_loss = total_loss / max(len(dataset), 1)
    print(f'Época {epoch+1}/{EPOCHS} — Loss: {media_loss:.4f}')

print('\n✅ Treino concluído!')
````

````python
# ETAPA 9 — Salvar LoRA treinado
import json

os.makedirs(OUTPUT_DIR, exist_ok=True)

# Salva o adapter LoRA
model.decoder.save_pretrained(OUTPUT_DIR)

# Salva o processor
processor.save_pretrained(OUTPUT_DIR)

# Cria README válido para o HF
with open(f'{OUTPUT_DIR}/README.md', 'w') as f:
    f.write('''---
base_model: facebook/musicgen-small
library_name: transformers
pipeline_tag: text-to-audio
tags:
- lora
- musicgen
- music-generation
- peft
---

# MusicGen LoRA Fine-Tuned

LoRA adapter fine-tuned on custom music dataset using `facebook/musicgen-small`.

## Como usar

```python
from transformers import AutoProcessor, MusicgenForConditionalGeneration
from peft import PeftModel
import torch, soundfile as sf

processor = AutoProcessor.from_pretrained("facebook/musicgen-small")
model = MusicgenForConditionalGeneration.from_pretrained("facebook/musicgen-small")
model.decoder = PeftModel.from_pretrained(model.decoder, "SEU_REPO_AQUI")

inputs = processor(text=["upbeat electronic music"], return_tensors="pt")
audio = model.generate(**inputs, max_new_tokens=256)
sf.write("output.wav", audio[0].cpu().numpy(), 32000)
```
''')

print('✅ LoRA salvo em:', OUTPUT_DIR)
print('Arquivos:')
for f in os.listdir(OUTPUT_DIR):
    tam = os.path.getsize(f'{OUTPUT_DIR}/{f}') / 1e6
    print(f'  {f}  ({tam:.1f} MB)')
````

````python
# ETAPA 10 — Gerar música com o modelo treinado
import torch, soundfile as sf
from IPython.display import Audio, display

MEU_PROMPT   = 'upbeat electronic dance music with heavy bass'  # ← seu prompt
DURACAO_GER  = 10   # segundos
AUDIO_SAIDA  = '/kaggle/working/musica_gerada.wav'

model.eval()
torch.cuda.empty_cache()

# Tokens por segundo no MusicGen = 50
max_tokens = DURACAO_GER * 50

print(f'🎵 Gerando: "{MEU_PROMPT}"...')
inputs = processor(text=[MEU_PROMPT], return_tensors='pt').to(DEVICE)

with torch.inference_mode():
    audio_values = model.generate(
        **inputs,
        max_new_tokens=max_tokens,
        do_sample=True,
        guidance_scale=3.0,
        temperature=1.0
    )

audio_np = audio_values[0, 0].cpu().float().numpy()
sf.write(AUDIO_SAIDA, audio_np, 32000)

print(f'✅ Música salva: {AUDIO_SAIDA}')
print(f'   Duração: {len(audio_np)/32000:.1f}s')

# Reproduz no Kaggle
display(Audio(audio_np, rate=32000))
````
